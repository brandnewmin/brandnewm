hypothesis & cost function : Andrew ng 1주차 강의 내용

```
x_train = [1,2,3]
y_train = [1,2,3]

w = tf.Variable(tf.random_normal([1]), name='weight')
#Variable : 텐서플로우의 변수 (기존 변수 개념이 아님, 텐서플로우가 변경시키는 값, trainable)
#[1] : 1차원 array
b = tf.Variable(tf.random_normal([1]), name='bias')
hypothesis = x_train*w + b 
```

```
cost = tf.reduce_mean(tf.square(hypothesis - y_train))
#tf.reduce_mean : 평균내주는 함수
```

```
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
train = optimizer.minimize(cost)
```

```
sess = tf.Session() #세션만들기
```

```
sess.run(tf.global_variables_initializer())
#Variable 만들었을 때 반드시 해줘야하는 작업(?)
```

```
for step in range(2001): #전체 스텝 2000번
  sess.run(train)
  if step % 20 == 0: #스텝 20번에 한번씩 출력하라
    print(step, sess.run(cost), sess.run(w), sess.run(b))
```

```
x = tf.placeholder(tf.float32)
#placeholder : x y 값을 미리 주지 않고, 필요할 때 제공한다.
y = tf.placeholder(tf.float32)
```

```
for steop in range(2001):
  _, cost_val, w_val, b_val = sess.run([train, cost, w, b], feed_dict = { x : [1,2,3,4,5], y : [2.1,3.1,4.1,5.1,6.1]}
                                                            #feed_dict는 앞에 x_train, y_train과 같은 기능이다

  if step % 20 == 0:
    print(step, cost_val, w_val, b_val)
```
